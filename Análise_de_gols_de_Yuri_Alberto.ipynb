### 5. Quais as faixas de tempo de todos os gols?


```python
# Normalização do campo "minuto" e criação dos ranges de tempo para a análise

# Remover o símbolo '+' e qualquer coisa que apareça após ele na coluna 'minuto_normalizado'
df['minuto_normalizado'] = df['minuto'].str.replace(r'\+.*$', '', regex=True)

# Definir a opção para exibir todas as linhas
pd.set_option('display.max_rows', None)

# Convertendo o campo 'minuto_normalizado' para o tipo inteiro
df['minuto_normalizado'] = df['minuto_normalizado'].astype(np.int64)

df.info()
```

```python
# Definir os rótulos para os ranges de tempo
def segmentar_faixa_tempo(valor):
    if valor <= 15:
        return '0-15'
    elif valor <= 30:
        return '15-30'
    elif valor <= 45:
        return '30-45'
    elif valor <= 60:
        return '45-60'
    elif valor <= 75:
        return '60-75'
    else:
        return '75-90'

# Aplicar a função aos valores da coluna 'minuto_normalizado' para criar os intervalos de tempo
df['intervalo_tempo'] = df['minuto_normalizado'].apply(segmentar_faixa_tempo)

# Contar as linhas em cada intervalo de tempo
contagem_por_intervalo = df['intervalo_tempo'].value_counts().sort_index()

print(contagem_por_intervalo)
```

```python
# Plotagem do gráfico de barras
plt.figure(figsize=(10, 6))
contagem_por_intervalo.plot(kind='bar', color='black')
plt.title('Contagem de Gols por Range de Tempo da partida')
plt.xlabel('Intervalo de Tempo')
plt.ylabel('Quantidade de gols')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

Yuri tende a marcar mais gols entre os 15 minutos finais de cada tempo e tende a fazer menos gols nos começos das partidas. Isso muito pode ser explicado devido as fortes marcações e estudos que o futebol moderno aplica na etapa inicial de cada partida, o que deixa o jogo mais truncado e com menos possibilidades de finalizações e gols.

### 7. Quais os resultados finais dos jogos em que Yuri fez gol? E quais os resultados parciais que os gols resultaram?
---



```python
# Função para extrair os gols do mandante
def extrair_gols_mandante(cidade, resultado):
    if cidade == 'F':
        return resultado.split(':')[0]
    elif cidade == 'C':
        return resultado.split(':')[0]
    else:
        return None

# Função para extrair os gols do visitante
def extrair_gols_visitante(cidade, resultado):
    if cidade == 'F':
        return resultado.split(':')[1]
    elif cidade == 'C':
        return resultado.split(':')[1]
    else:
        return None

# Aplicar as funções às colunas 'cidade' e 'resultado' para criar as novas colunas
df['gols_mandante'] = df.apply(lambda x: extrair_gols_mandante(x['cidade'], x['resultado']), axis=1)
df['gols_visitante'] = df.apply(lambda x: extrair_gols_visitante(x['cidade'], x['resultado']), axis=1)

# Função para determinar o resultado final do jogo
def determinar_resultado_final(cidade, gols_mandante, gols_visitante):
    if cidade == 'C':
        if gols_mandante > gols_visitante:
            return 'vitória'
        elif gols_mandante < gols_visitante:
            return 'derrota'
        else:
            return 'empate'
    elif cidade == 'F':
        if gols_mandante > gols_visitante:
            return 'derrota'
        elif gols_mandante < gols_visitante:
            return 'vitória'
        else:
            return 'empate'
    else:
        return None

# Aplicar a função às colunas 'cidade', 'gols_mandante' e 'gols_visitante' para criar a nova coluna
df['resultado_final'] = df.apply(lambda x: determinar_resultado_final(x['cidade'], x['gols_mandante'], x['gols_visitante']), axis=1)


# Contar o número de ocorrências de cada categoria em 'resultado_parcial'
contagem_resultado_final = df['resultado_final'].value_counts()

# Plotar o gráfico de barras
contagem_resultado_final.plot(kind='bar', color='black')

# Adicionar rótulos e título ao gráfico
plt.xlabel('Resultado Final')
plt.ylabel('Quantidade de gols')
plt.title('Quantidade de gols por Resultado Final')
plt.xticks(rotation=90)
# Exibir o gráfico
plt.show()
```

```python
# Função para determinar o resultado parcial do jogo
def determinar_resultado_parcial(resultado, para_resultado):
    if para_resultado == resultado:
        return 'fechou o placar'
    elif para_resultado == '0:1' or para_resultado == '1:0':
        return 'abriu o placar'
    elif para_resultado == '1:1' or para_resultado == '2:2' or para_resultado == '3:3' or para_resultado == '4:4' or para_resultado == '5:5' or para_resultado == '6:6':
        return 'empatou'
    else:
        return 'ampliou ou diminuiu o placar'

# Aplicar a função às colunas 'resultado' e 'para_resultado' para criar a nova coluna 'resultado_parcial'
df['resultado_parcial'] = df.apply(lambda x: determinar_resultado_parcial(x['resultado'], x['para resultado']), axis=1)

# Contar o número de ocorrências de cada categoria em 'resultado_parcial'
contagem_resultado_parcial = df['resultado_parcial'].value_counts()

# Plotar o gráfico de barras
contagem_resultado_parcial.plot(kind='bar', color='black')

# Adicionar rótulos e título ao gráfico
plt.xlabel('Resultado Parcial')
plt.ylabel('Quantidade de gols')
plt.title('Quantidade de gols por Resultado Parcial')
plt.xticks(rotation=90)
# Exibir o gráfico
plt.show()
```

É evidente que os jogos que o atacante marca gol são finalizados com vitória do time do jogador.

Também é importante destacar que a grande volumetria dos gols de Yuri são do tipo "ampliou ou diminuiu o placar" e "fechou o placar", mostrando que o atacante conta com boa conversão durante o andamento da partida.

### 8. Quantos minutos para participar de um gol Yuri Alberto tem por temporada?





```python
# Contar o número de ocorrências de cada categoria em 'resultado_parcial'
df2['participacoes'] = df2['gols'] + df2['assistencias']

# Indicador de participações por minuto
df2['participacoes_por_minuto'] = df2['minutos_jogados'] / df2['participacoes']
df2['participacoes_por_minuto'] = df2['participacoes_por_minuto'].where(df2['participacoes'] != 0, 0).round(2)
df2

# Agrupar os dados por ano e calcular a média das participações por minuto
media_por_ano = df2.groupby('temporada')['participacoes_por_minuto'].mean()

# Plotar o gráfico de linha
plt.figure(figsize=(10, 6))
media_por_ano.plot(kind='line', color='black', marker='o', linestyle='-')
plt.title('Minutos para participar de um gol por temporada')
plt.xlabel('Ano')
plt.ylabel('Minutos médios para participar de um gol')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 9. Qual competição Yuri tem o menor tempo médio para participar de um gol?





```python
# Agrupar os dados por ano e calcular a média das participações por minuto
media_por_campeonato = df2.groupby('competicao')['participacoes_por_minuto'].mean()

# Plotar o gráfico de linha
plt.figure(figsize=(10, 6))
media_por_campeonato.plot(kind='line', color='black', marker='o', linestyle='-')
plt.title('Minutos para participar de um gol por competição')
plt.xlabel('Competição')
plt.ylabel('Minutos médios para participar de um gol')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
